```python
t=[]
for i in range(0, len(data)):
  sample = data.iloc[i, 3]
  t.append(sample)
```

```python
len(t)
```

```python
data.iloc[0, 3]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()

tfidf = vectorizer.fit_transform(t)
```

```python
tfidf.shape
```

```python
print(tfidf.toarray())
```

```python
from gensim import models, matutils

corpus = matutils.Sparse2Corpus(tfidf, documents_columns=False)
```

```python
from gensim.corpora import Dictionary

id2word = Dictionary.from_corpus(corpus, id2word=id2word)
```

```python
lda = models.LdaModel(corpus=corpus, num_topics=5, id2word=id2word, passes=10)
```
