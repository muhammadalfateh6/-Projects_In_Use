```python
import docx2txt
import nltk
from nltk.corpus import stopwords
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.corpora.dictionary import Dictionary
from gensim.models.ldamodel import LdaModel
import pyLDAvis
import pyLDAvis.gensim
```

```python
text = docx2txt.process("/content/prophets.docx")
print(text)
```

```python
stop_words = stopwords.words('arabic')
stop_words
```

```python
text_2 = text[0: 100]
```

```python
text_2
```

```python
new = simple_preprocess(text)
print(new)
```

```python
new_1 = [d.split() for d in new]

dic = Dictionary(new_1)
```

```python
dic[[5][:1][0]]
```

```python
print(dic)
```

```python
my_bow = [dic.doc2bow(new),]
# corpus2 = [dct.doc2bow(filtered_sentence),]
```

```python
print(my_bow)
```

```python
model = LdaModel(corpus=my_bow,
                 id2word=dic,
                 num_topics=5,
                 random_state=100,
                 update_every=1,
                 chunksize=50,
                 passes=10,
                 alpha='auto')
```

```python
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(model, my_bow, dic, mds='mmds', R=30)
```
